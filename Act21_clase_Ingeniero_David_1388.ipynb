```python
class Ingeniero:
    # Atributos actualizados (incluye el proyecto)
    def __init__(self, nombre, especialidad, experiencia, proyecto):
        self.nombre = nombre
        self.especialidad = especialidad
        self.experiencia = experiencia
        self.proyecto = proyecto

    # Funciones / Métodos
    def diseñar(self):
        return f"📐 {self.nombre} está diseñando los planos para: '{self.proyecto}'."

    def supervisar(self):
        return f"👷 {self.nombre} está en el sitio de '{self.proyecto}' supervisando la ejecución."

    def calcular(self):
        return f"⌨️ {self.nombre} realiza cálculos estructurales para el proyecto '{self.proyecto}'."

# --- Entrada de datos interactiva ---

print("--- SISTEMA DE GESTIÓN DE PROYECTOS DE INGENIERÍA ---")

# Registro del Ingeniero 1
print("\n>>> Datos del Ingeniero 1:")
n1 = input("Nombre: ")
e1 = input("Especialidad: ")
ex1 = input("Años de experiencia: ")
p1 = input("Nombre del proyecto asignado: ")
ingeniero1 = Ingeniero(n1, e1, ex1, p1)

# Registro del Ingeniero 2
print("\n>>> Datos del Ingeniero 2:")
n2 = input("Nombre: ")
e2 = input("Especialidad: ")
ex2 = input("Años de experiencia: ")
p2 = input("Nombre del proyecto asignado: ")
ingeniero2 = Ingeniero(n2, e2, ex2, p2)

# --- Presentación de resultados ---

print("\n" + "="*40)
print("ESTADO ACTUAL DE LOS PROYECTOS")
print("-" * 40)

# Acciones del objeto ingeniero1
print(f"\nINGENIERO: {ingeniero1.nombre}")
print(f"Especialidad: {ingeniero1.especialidad}")
print(ingeniero1.diseñar())
print(ingeniero1.supervisar())

# Acciones del objeto ingeniero2
print(f"\nINGENIERO: {ingeniero2.nombre}")
print(f"Especialidad: {ingeniero2.especialidad}")
print(ingeniero2.calcular())
print(ingeniero2.diseñar())

print("\nPrograma creado por David Barrera 1388")
```
